### *NOMOR 1*

```python
kata = input("Masukkan kata: ")
class Stack:                                                       #1
    StackKata = []
    
    def __init__(self, kata):                                      #2
        self.kata = kata
    def push(self):                                                #3
        for char in self.kata:
            self.StackKata.append(char)
    def pop(self):                                                 #4
        hasil = ""
        self.push()
        if len(self.StackKata) > 0:
            for i in reversed(range(len(self.StackKata))):
                hasil += self.StackKata[i]
            print("Kata yang sudah dibalik: ", hasil)
        else:
            print("Maaf stack masih kosong!")
        
if __name__ == "__main__":                                         #5
    Stack(kata).pop()
```

```python
class Queue:                                                                            #1
    data_antrian = input('Masukkan data antrian: ')
    daftar_peserta = data_antrian.split(',')
    daftar_antrian = []
    max_antrian = int(input('Jumlah Antrian: '))
    
    def enqueue(self):                                                                  #2
        for i in range(len(self.daftar_peserta)):
            if i >= self.max_antrian:
                print('Maaf', str(self.daftar_peserta[i])+', antrian sudah penuh')
                continue
            else:
                print(str(self.daftar_peserta[i]), 'masuk antrian')
                self.daftar_antrian.extend(self.daftar_peserta[i])
        print('=' * 10)
        
    def dequeue(self):                                                                  #3
        self.enqueue()
        for i in range(len(self.daftar_antrian)):
            print(self.daftar_antrian[0], 'keluar antrian')
            self.daftar_antrian.pop(0)
            for nama in self.daftar_antrian:
                print(str(nama)+', ', end='')
            
            #print sisa antrian (yg kosong)
            kosong = self.max_antrian - len(self.daftar_antrian)
            print('kosong, ' * kosong)
            print('=' * 10)
        
        print('Antrian Kosong')
        print('Jumlah Antrian:', len(self.daftar_antrian))
        
if __name__=='__main__':                                                                #4
    q = Queue()
    q.dequeue()
```

### *NOMOR 3*

```python
class Node:                                                                                 #1
    def __init__(self, val):
        self.left = None
        self.mid = None
        self.right = None
        self.val = val

def preorder_traversal(node):                                                               #2
    if node:
        print(node.val, end=' ')
        preorder_traversal(node.left)
        preorder_traversal(node.mid)
        preorder_traversal(node.right)

def inorder_traversal(node):                                                                #3
    if node:
        inorder_traversal(node.left)
        print(node.val, end=' ')
        inorder_traversal(node.mid)
        inorder_traversal(node.right)

def postorder_traversal(node):                                                              #4
    if node:
        postorder_traversal(node.left)
        postorder_traversal(node.mid)
        postorder_traversal(node.right)
        print(node.val, end=' ')

def get_leaves(node):                                                                       #5
    if node is None:
        return []
    leaves = []
    if node.left is None and node.mid is None and node.right is None:
        return [node.val]
    if node.left:
        leaves += get_leaves(node.left)
    if node.mid:
        leaves += get_leaves(node.mid)
    if node.right:
        leaves += get_leaves(node.right)
    return leaves

if __name__ == '__main__':                                                                  #6
    root = Node('Z')
    root.left = Node('A')
    root.left.left = Node('X')
    root.left.left.left = Node('B')
    root.left.mid = Node('N')
    root.left.right = Node('W')
    root.left.right.left = Node('Q')
    root.mid = Node('G')
    root.right = Node('R')
    root.right.right = Node('F')
    root.right.right.left = Node('S')
    root.right.right.left.left = Node('D')
    root.right.right.left.right = Node('K')
    root.right.right.right = Node('J')
    root.right.right.right.right = Node('P')
    
    print('Preorder Traversal: ', end='')                                                   #7
    preorder_traversal(root)
    print()

    print('Inorder Traversal: ', end='')
    inorder_traversal(root)
    print()

    print('Postorder Traversal: ', end='')
    postorder_traversal(root)
    print()

    print('Leaf: ', end='')
    print(' '.join(get_leaves(root)))

    print('Root: ', end='')
    print(root.val)
```
