### *NOMOR 1*

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

#### melakukan import modul

```python
dataregresi = pd.read_csv('SeoulBikeData.csv', sep=',')
X_train = dataregresi['Temperature'].values.reshape(-1, 1)
y_train = dataregresi['Rented Bike Count'].values
```

#### melakukan import data csv lalu di read dan diklasifikasikan ke dalam x dan y. reshape(-1, 1) untuk mengubah bentuk agar menjadi urutan kebawah

```python
modelregresilinier = LinearRegression()
modelregresilinier.fit(X_train, y_train)
```

#### menginstruksikan model regresi linear pada variabel X_train dan y_train

```python
plt.scatter(X_train, y_train, color='k')
plt.plot(X_train, modelregresilinier.predict(X_train), color='r')

plt.show()
```

#### membuat scatter plot dengan variabel X_train dan y_train. menggunakan parameter warna hitam untuk titik-titik dan warna merah untuk garis plot

```python
TemperatureKota = [[20], [25], [30], [15], [17]]
JumlahPenyewa = modelregresilinier.predict(TemperatureKota)
```

#### list berisi temperature kota yang akan diprediksi. selanjutnya memprediksi nilai jumlah penyewa dengan model regresi linear yang sudah ada dan berdasarkan nilai pada list temperature kota 

```python
for Temperature, prediksi in zip(TemperatureKota, JumlahPenyewa ):
    print(f'Temperature = {Temperature[0]} celcius, Prediksi Jumlah Penyewa Sepeda = {prediksi}')
```

#### melakukan looping dengan zip untuk mengiterasi list temperature kota dan jumlah penyewa. selanjutnya untuk menghasilkan output dibutuhkan f-string untuk menggabungkan nilai-nilai variabel ke dalam string
