#### melakukan import

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
from sklearn import datasets
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay
```

#### melakukan import data

```python
dataset = pd.read_csv('Iris_Data.csv')
```

#### mengeluarkan species dari variabel x dan memasukkan menjadi variabel y

```python
X = dataset.drop('species', axis=1)
Y = dataset['species']
```

#### membagi data test dengan data testing dengan test size 0.3 sebagai subset pengujian dan random state 100 untuk mengatur seed atau nilai awal yang digunakan dalam proses pembagian data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

#### menamai x.columns dan y.columns untuk memudahkan interpretasi dan penggunaan data dalam analisis

```python
X.columns = ['sepal-lenght', 'sepal-width', 'petal-length', 'petal-width']
Y.columns = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

#### membuat model decision tree dengan random state 100 agar hasil pembentukan model menjadi deterministik dan dapat direproduksi

```python
clf= DecisionTreeClassifier(random_state = 100)
```

#### memasukkan data training ke dalam model decision tree yang disimpan dalam variabel clf dan menghasilkan keputusan yang digunakan untuk melakukan prediksi pada data baru

```python
clf.fit(X_train, Y_train)
```

#### melakukan prediksi terhadap data testing dan menghasilkan output dari variabel y_pred

```python
Y_pred = clf.predict(X_test)
print(Y_pred)
```

#### mencetak nilai akurasi dari model dengan accuracy_score dari modul metrics digunakan untuk menghitung akurasi berdasarkan variabel y_test dan y_pred. lalu dilakukan confusion_matrix dengan variabel y_test dan y_pred untuk menunjukkan jumlah klasifikasi yang benar dan salah dari setiap kelas. dan dilalukan print untuk menghasilkan output

```python
print('Accuracy:', metrics.accuracy_score(Y_test, Y_pred))
cm = confusion_matrix(Y_test, Y_pred)
print('===CONFUSION MATRIX===')
print(cm)
```

####  menampilkan grafik dengan membuat objek gambar (figure) dengan nomor identifikasi 1. lalu mebentuk gambar dengan parameter clf. kemudian menghasilkan visualisasi dengan plot_tree. 

```python
figl = plt.figure(1)
tree.plot_tree(clf, feature_names=X.columns, class_names=Y.columns)
display(figl)
```

#### menampilkan grafik dengan membuat objek gambar (figure) dengan nomor identifikasi 2. lalu membuat heatmap berdasarkan confusion matrix (cm). annot=True untuk menampilkan nilai-nilai cm dalam heatmap. xticklabels=Y.columns dan yticklabels=Y.columns untuk memberikan label pada sumbu x dan y.

```python
fig2 = plt.figure(2)
sns.heatmap(cm, annot=True, xticklabels=Y.columns, yticklabels=Y.columns)
display(fig2)
```
